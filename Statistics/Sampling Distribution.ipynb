```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weight-height.csv')
df
```

```python
#population mean
df['Height'].mean()
```

```python
s = df['Height'].sample(n = 50, random_state=1)
```

```python
#sample mean
s.mean()
```

```python
sample_means = []
for i in range(1, 1000):
    s = df['Height'].sample(n = 50)
    sample_means.append(s.mean())
sample_means
```

```python
#sample plot
plt.hist(sample_means, bins = 20)
plt.show()
```

```python
#population plot
plt.hist(df['Height'], bins = 20)
plt.show()
```
