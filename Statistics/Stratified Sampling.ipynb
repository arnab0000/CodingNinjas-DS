```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('startup_funding.csv', encoding = 'utf-8')
data.head()
```

```python
data["CityLocation"].replace('bangalore', 'Bangalore', inplace = True)
data["CityLocation"].replace('Delhi', 'New Delhi', inplace = True)
```

```python
data = data[data['AmountInUSD'].notna()]
```

```python
data['AmountInUSD'] = data['AmountInUSD'].str.replace(",", "")
```

```python
data['AmountInUSD'] = pd.to_numeric(data['AmountInUSD'])
```

```python
data['AmountInUSD'].dropna(inplace = True)
```
